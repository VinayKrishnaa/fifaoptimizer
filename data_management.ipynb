# Step 1: Make Data searchable by position name

```python

# Extract the player positions column
positions = data['player_positions']

# Create a list of unique positions
unique_positions = []
for pos_list in positions:
    for pos in pos_list.split(','):
        if pos.strip() not in unique_positions:
            unique_positions.append(pos.strip())

# Create a new dataframe with one-hot encoded positions
position_df = pd.DataFrame(0, index=data.index, columns=unique_positions)

# Iterate through the original positions and set the corresponding columns to 1
for i, pos_list in enumerate(positions):
    for pos in pos_list.split(','):
        position_df.loc[i, pos.strip()] = 1

# Concatenate the original dataframe with the one-hot encoded positions
data_plus_positions = pd.concat([data, position_df], axis=1)

# Print the updated dataframe
print(data_plus_positions.head())

data_plus_positions
```

```python
import csv

# Read the dataset into a list of dictionaries
players = []
with open('paste.txt') as file:
    reader = csv.DictReader(file, delimiter='\t')
    for row in reader:
        players.append(row)

# Function to find a player by short name
def find_player(short_name):
    for player in players:
        if player['short_name'] == short_name:
            return player
    return None

# Function to extract a team of 11 players
def extract_team(short_names):
    team = []
    for short_name in short_names:
        player = find_player(short_name)
        if player:
            team.append(player)
        else:
            print(f"Player with short name '{short_name}' not found.")
    return team

# Prompt the user for player short names
player_names = []
print("Enter the short names of 11 players (or 'quit' to finish):")
while len(player_names) < 11:
    short_name = input(f"Player {len(player_names) + 1}: ").strip()
    if short_name.lower() == 'quit':
        break
    player_names.append(short_name)

# Extract the team
team = extract_team(player_names)

# Print the extracted team
print("\nExtracted Team:")
for player in team:
    print(f"{player['short_name']} - {player['long_name']} ({player['player_positions']})")
```
