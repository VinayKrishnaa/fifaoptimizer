### Functions to select players from Dataset

#### Function 1: Select players randomly based on formation

```python
def select_players(df, formation):
    
    selected_player_ids = []
    # Get the position requirements for the given formation
    position_requirements = get_position_requirements(formation)
    
    
    if position_requirements == "Invalid formation":
        return "Invalid formation"
    
    # Create a new DataFrame to store the selected players
    selected_players = pd.DataFrame(columns=df.columns)
    
    for position, count in position_requirements.items():
        # Sort the players by the position column in descending order
        sorted_players = df.sort_values(by=position, ascending=False)
        
        # Exclude already selected players
        sorted_players = sorted_players[~sorted_players['sofifa_id'].isin(selected_player_ids)]
        
        # Select the top `count` players for the current position
        selected_position_players = sorted_players.loc[sorted_players[position] == 1].head(count)
        
        # Append the selected players to the `selected_players` DataFrame
        selected_players = pd.concat([selected_players, selected_position_players])
        
        # Update the list of selected player IDs
        selected_player_ids.extend(selected_position_players['sofifa_id'])
    
    # Return the selected players DataFrame
    return selected_players
```

#### Function 2: Select given number of players by Nationality
Typical method of Game Start on Fifa

```python
import random
import pandas as pd

def select_players_nationality(players, nationality, number):
    selected_players = pd.DataFrame()  # Initialize an empty DataFrame

    # Filter players by nationality
    nationality_players = players[players['nationality_name'] == nationality]

    # Check if the requested number is greater than the available players
    if number > len(nationality_players):
        print(f"Not enough players available for nationality {nationality}.")
        return []

    # Randomly select players
    selected_indices = random.sample(range(len(nationality_players)), number)

    # Add selected players to the result DataFrame
    selected_players = pd.concat([selected_players, nationality_players.iloc[selected_indices]])

    return selected_players
```

### Function to allocate best player to a particular position

```python
def allocate_players(selected_players, position_requirements):
    allocated_players = []
    allocated_player_names = set()

    # Iterate through each position and the number of players needed for that position
    for position, count in position_requirements.items(): #GK
        position_candidates = selected_players[selected_players[position] == 1]
        position_lower = position.lower()#gk
        print(position)
        
       # Sort candidates by position score in descending order
        sorted_candidates = position_candidates.sort_values(by=position_lower, ascending=False)
        
        selected_position_players = []
        for _, player in sorted_candidates.iterrows():
            player_name = player['short_name']

            if player_name not in allocated_player_names:
                
                selected_position_players.append(player)
                allocated_player_names.add(player_name)

            if len(selected_position_players) == count:
                break
        
        for player in selected_position_players:
             allocated_players.append({
                'sofifa_id': player['sofifa_id'],
                'Short Name': player['short_name'],
                'Allocated Position': position,
                'Position Score': int(player[position_lower])
             })

    # Create DataFrame from the result list
    allocated_players_table = pd.DataFrame(allocated_players)
    
    return allocated_players_table  
```

### Function to calculate chemistry of given 11 player team

```python
def player_chemistry(selected_11_players):
    # Initialize nationality score column
    selected_11_players['Nationality Score'] = 0
    
    # Calculate nationality score based on conditions
    for country in selected_11_players['nationality_name'].unique():
        country_players = selected_11_players[selected_11_players['nationality_name'] == country]
        num_players = len(country_players)
        if num_players >= 2:
            selected_11_players.loc[selected_11_players['nationality_name'] == country, 'Nationality Score'] = 1
        if num_players >= 5:
            selected_11_players.loc[selected_11_players['nationality_name'] == country, 'Nationality Score'] = 2
        if num_players >= 8:
            selected_11_players.loc[selected_11_players['nationality_name'] == country, 'Nationality Score'] = 3
    
    # Initialize league score column
    selected_11_players['League Score'] = 0
    
    # Calculate league score based on conditions
    for country in selected_11_players['league_name'].unique():
        country_players = selected_11_players[selected_11_players['league_name'] == country]
        num_players = len(country_players)
        if num_players >= 3:
            selected_11_players.loc[selected_11_players['league_name'] == country, 'League Score'] = 1
        if num_players >= 5:
            selected_11_players.loc[selected_11_players['league_name'] == country, 'League Score'] = 2
        if num_players >= 8:
            selected_11_players.loc[selected_11_players['league_name'] == country, 'League Score'] = 3
            
     # Initialize team score column
    selected_11_players['Club Score'] = 0
    
    # Calculate league score based on conditions
    for country in selected_11_players['club_name'].unique():
        country_players = selected_11_players[selected_11_players['club_name'] == country]
        num_players = len(country_players)
        if num_players >= 3:
            selected_11_players.loc[selected_11_players['club_name'] == country, 'Club Score'] = 1
        if num_players >= 5:
            selected_11_players.loc[selected_11_players['club_name'] == country, 'Club Score'] = 2
        if num_players >= 8:
            selected_11_players.loc[selected_11_players['club_name'] == country, 'Club Score'] = 3
             
    selected_11_players["Chemistry Score"] = selected_11_players[['Nationality Score', 'League Score', 'Club Score']].max(axis=1)
        
    
    return selected_11_players["Chemistry Score"].sum()
```

```python
import gurobipy as gp
from gurobipy import GRB

def optimize_player_allocation(players_array, formations_array):
    # Get the dimensions of the players array
    num_players = len(players_array)
    num_positions = len(players_array[0])
    num_formations = len(formations_array)

    # Initialize the maximum total score and the best formation
    max_total_score = 0
    best_formation = None

    # Loop through each formation
    for formation in formations_array:
        # Create the Gurobi model
        model = gp.Model()

        # Create the decision variables
        x = model.addVars(num_players, num_positions, vtype=GRB.BINARY, name="player_allocation")

        # Set the objective function
        objective = gp.quicksum(players_array[i][j] * x[i, j] for i in range(num_players) for j in range(num_positions))
        model.setObjective(objective, GRB.MAXIMIZE)

        # Add the constraints
        # 1. Each player can only play in one position
        for i in range(num_players):
            model.addConstr(gp.quicksum(x[i, j] for j in range(num_positions)) == 1)

        # 2. The number of players in each position must match the formation
        for j in range(num_positions):
            model.addConstr(gp.quicksum(x[i, j] for i in range(num_players)) == formation[j])

        # Solve the optimization problem
        model.optimize()

        # Retrieve the solution
        player_positions = []
        for i in range(num_players):
            for j in range(num_positions):
                if x[i, j].X > 0.5:
                    player_positions.append(j)
                    break

        # Calculate the total player score
        total_score = 0
        for i in range(num_players):
            total_score += players_array[i][player_positions[i]]

        # Update the maximum total score and the best formation if necessary
        if total_score > max_total_score:
            max_total_score = total_score
            best_formation = formation

    return best_formation, max_total_score
```
